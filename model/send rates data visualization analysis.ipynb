# Import library

```python
import pandas as pd
```

# Data visualization analysis

```python
df = pd.read_csv('../data/transaction_send_rates/ReserveSendRates_5minutes.csv')
df.head()
```

```python
df.shape
```

```python
df.index = pd.to_datetime(df['Time'])
```

```python
df.head()
```

```python
df = df.drop('Time', axis=1)
df.head()
```

```python
import seaborn as sns
from pylab import rcParams
import matplotlib.pyplot as plt
from matplotlib import rc

%matplotlib inline
%config InlineBackend.figure_format='retina'

sns.set(style='whitegrid', palette='muted', font_scale=1.2)

HAPPY_COLORS_PALETTE = ["#01BEFE", "#FFDD00", "#FF7D00", "#FF006D", "#93D30C", "#8F00FF"]

sns.set_palette(sns.color_palette(HAPPY_COLORS_PALETTE))

rcParams['figure.figsize'] = 24, 10

from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
plt.plot(df[0:8640])
plt.title('Number of transaction send rates every 5 minute')
```

```python
# Save transaction occurrence data as npy file
import numpy as np


Number = df['Number'].values.reshape(8640, 1)
np.save("../data/transaction_send_rates/ReserveSendRates_5minutes.npy", Number)
```
